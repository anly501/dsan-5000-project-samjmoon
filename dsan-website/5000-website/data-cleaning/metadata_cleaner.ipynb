```python
import pandas as pd
import numpy as np
from ast import literal_eval
import os
os.chdir('/Users/samanthamoon/Desktop/DSAN 5000/dsan-5000-project-samjmoon/dsan-website/5000-website/data-cleaning')

metadata_df = pd.read_csv('movies_metadata.csv')
metadata_df['genres'] = metadata_df['genres'].apply(lambda x: [genre['name'] for genre in literal_eval(x)] if isinstance(x, str) else [])
metadata_df.head()
```

```python
num_columns = metadata_df.shape[1]
num_columns
column_names = metadata_df.columns.tolist()
print("Column names:", column_names)
num_rows = metadata_df.shape[0]
num_rows
```

```python
columns_to_drop = ['adult','belongs_to_collection','homepage', 'overview', 'poster_path', 'tagline', 'video', 'status', 'production_countries', 'original_language', 'spoken_languages', 'production_companies']
metadata_df = metadata_df.drop(columns=columns_to_drop, axis=1)
metadata_df.head()
metadata_df = metadata_df.rename(columns={'id': 'tmdbId'})
metadata_df['tmdbId'] = pd.to_numeric(metadata_df['tmdbId'], errors='coerce')
metadata_df['tmdbId'] = metadata_df['tmdbId'].astype(float)
#Check if any column in the DataFrame contains NaN values
any_nan_values = metadata_df.isna().any()

#Display which columns have NaN values
print(any_nan_values)
```

```python
metadata_df = metadata_df.dropna(subset=['tmdbId', 'imdb_id', 'vote_average', 'vote_count', 'runtime'])
num_rows = metadata_df.shape[0]
num_rows
any_nan_values = metadata_df.isna().any()

# Display which columns have NaN values
print(any_nan_values)
metadata_df.head()
```

```python
file_path = './cleaned_metadata_movies.csv' 
metadata_df.to_csv(file_path, index=False) 
```
